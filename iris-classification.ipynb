```python
#Importing all the needed Libraries 
import pandas as pd #pandas for data Handling
import numpy as np # numpy for numerical Operations
import matplotlib.pyplot as plt #Matplotlib for Plotting Graphs
import seaborn as sns #seaborn for advanced vizualization
#scikit-learn for ML models
#ML part:->
from sklearn.datasets import load_iris  #provides built-in datasets
from sklearn.model_selection import train_test_split #splits the data into training data and testing data
from sklearn.linear_model import LogisticRegression #it is a classification algorithm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix #all of these are Evalution Tools
```

```python
#Load Your Dataset
df = pd.read_csv("Iris.csv") #Reads CSV file and converts it into to a table
df.head() #df = DataFrame(table-like structure)
```

```python
#Check Structures
df.info() #Shows no of rows, column names, data types, missing values
df.describe() #give mean, min, max, S.D
df.columns #Display all column names
```

```python
#Data Cleaning
if "Id" in df.columns:
  df = df.drop("Id", axis=1) #drop() - removes columns, "Id" = Column Name, axis = 1 - means we dropping a column
```

```python
#Checking Missing values
df.isnull().sum()
#isnull() = checks for missing values
#sum() = counts how many missing values per column
```

```python
#If any missing values appear:
df = df.dropna()
```

```python
#EDA - Exploratory Data Analysis
sns.countplot(x='species', data=df) #shows how many flowers of each species exist
plt.show()
```

```python
#pairplot - creates scatterplots btw every feature combination
#hue='species' -> colors the graph based on species
sns.pairplot(df, hue='species') 
plt.show()
```

```python
#Prepare Features and Target
X = df.drop('species', axis=1) #input features(measurements)
Y = df['species'] #output(target variable)
```

```python
#Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
X,Y,
test_size = 0.2,
random_state = 42 #Ensure same split every time
)
#20% data for testing 
#80% for training
```

```python
#Train Model
model = LogisticRegression(max_iter=200)
model.fit(X_train, Y_train)
# X_train = training features
# X_test = testing featues
# Y_test = testing features
# Y_train = training features
```

```python
#Make Prediction 
Y_pred = model.predict(X_test)
```

```python
#Evaluate Model
accuracy = accuracy_score(Y_test, Y_pred)
print("accuracy: ", accuracy)
```

```python
#Classification Report
print(classification_report(Y_test,Y_pred))
```

```python
#Confusion Matrix
cm = confusion_matrix(Y_test, Y_pred)

sns.heatmap(cm, annot=True, cmap="Blues") #converts confusion matrix into colored grid
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)

dt_pred = dt.predict(X_test)

print("Decision Tree Accuracy: ", accuracy_score(Y_test, dt_pred))
```
